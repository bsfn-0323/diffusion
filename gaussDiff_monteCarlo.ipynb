```python
import numpy as np
import matplotlib.pyplot as plt
import torch 
```

```python
def load_data(L,temp,MCS):
    return np.fromfile(f"../ising_wolff/dataIsing2D_L{L}/config_L{L}_T{temp:.3f}.bin",dtype = np.int32).reshape(MCS,L**2)
```

```python
Tmin = 2.27
Tmax = 3.22
meas = 20
Ts = np.linspace(Tmin,Tmax,meas)
L = 8
N = L**2
#data = load_data(L,Ts[9],200000)
data = load_data(L,Ts[12],200000)
plt.hist(np.mean(data,axis = 1))
```

Get interaction matrix and g

```python
C0 = np.cov(data.T)
invC0 = np.linalg.inv(C0)
num = 3*np.trace(invC0)
den = np.trace(np.matmul(invC0,np.matmul(C0+3*np.eye(N),C0+3*np.eye(N))))
g = -num/den
print(f"g = {g:.4f}")
```
